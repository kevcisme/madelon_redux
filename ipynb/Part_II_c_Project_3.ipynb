# Ten Percent

For the following models, I used a dataset of 22,200 rows and 5,000 columns. This is equivalent to 10% (ten percent) of my initial dataset of 220,000 rows and 5,000 columns.

#### Import the Libraries

```python
from os import chdir
chdir('/Users/mandymoody/DSI/projects/project_3')

%matplotlib inline

from lib import *
```

```python
from datetime import datetime
from IPython.display import display
from sklearn.decomposition import PCA 
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier 
from sklearn.feature_selection import SelectKBest, SelectFromModel, f_classif
from sklearn.linear_model import LogisticRegression, Lasso
from sklearn.model_selection import GridSearchCV, ShuffleSplit
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from scipy import stats, mean 
```

```python
feat22000_df = pd.read_pickle('22000feature_df.p')
target22000_df = pd.read_pickle('22000target_df.p')
```

#### Get an Array of Values

```python
X = feat22000_df.values
y = target22000_df.values
```

#### Decide on a pipeline for data to pass through

```python
seventh_pipeline = Pipeline([
    ('pca', PCA()),
    ('skb', SelectKBest(k=40)),
    ('rf', RandomForestClassifier())
])
```

```python
seventh_pipe_params = {
    'rf__n_estimators':[10, 40, 100],
    'rf__max_depth':[10,40,None]
}
```

#### GridSearchCV that pipeline

```python
seventh_pipeline_gs = GridSearchCV(seventh_pipeline,
                                 param_grid=seventh_pipe_params, 
                                 n_jobs=-1,
                                 cv=ShuffleSplit(random_state=42)                               
)
```

#### Fit it. Clock the total time

```python
seventh_start_time = datetime.now()
seventh_pipeline_gs.fit(X, y.ravel())
seventh_end_time = datetime.now()
```

```python
seventh_total_time = seventh_end_time - seventh_start_time
total_minutes_22k_seventh_pipeline = (seventh_total_time.total_seconds()) / 60
total_minutes_22k_seventh_pipeline
```

####  Load the results into a Dataframe to display results (using cv_results)

```python
seventh_pipeline_results = pd.DataFrame(seventh_pipeline_gs.cv_results_)
seventh_pipeline_results.T
```

#### Pickle the Dataframe for easy access in presentation

```python
seventh_pipeline_results.to_pickle('seventh_pipeline_results22k.p')
```
